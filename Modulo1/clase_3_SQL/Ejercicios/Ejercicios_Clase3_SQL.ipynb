## 📋 Referencia rápida

| Cláusula | Uso |
|----------|-----|
| `SELECT col AS alias` | Seleccionar y renombrar columnas |
| `DISTINCT` | Eliminar duplicados |
| `WHERE condición` | Filtrar filas |
| `AND / OR / NOT` | Combinar condiciones |
| `BETWEEN a AND b` | Rango inclusivo |
| `IN (v1, v2)` | Lista de valores |
| `ORDER BY col DESC` | Ordenar resultado |
| `LIMIT n` | Limitar filas devueltas |
| `GROUP BY col` | Agrupar filas |
| `COUNT / SUM / AVG / MAX / MIN` | Funciones de agregación |
| `HAVING condición` | Filtrar grupos (post-agrupación) |
| `INNER JOIN t ON a = b` | Solo filas con coincidencia en ambas tablas |
| `LEFT JOIN t ON a = b` | Todas las filas de la izquierda + coincidencias |
| `COALESCE(val, default)` | Reemplazar NULL por un valor por defecto |
| `SUBSTR(col, inicio, largo)` | Extraer parte de un texto |
| `ROUND(valor, decimales)` | Redondear un número |
