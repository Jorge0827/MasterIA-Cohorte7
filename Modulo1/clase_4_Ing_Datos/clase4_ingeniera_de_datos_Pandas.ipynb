## Tratando nulos

```python
#Regla de oro: Nunca elimines un nulo sin enteder por qué está ahí!!!!!!!!!!!

#Buena practica: Siempre trabaja sobre una copia para no dañar el archivo orginal

#Creación de copias.
df_ventas_1 = df_ventas.copy()
df_productos_1 = df_productos.copy()
df_clientes_1 = df_clientes.copy()
```

```python
#Ventas: cantidad nulas --> asumir 1 unidad (minimo logico)

df_ventas_1['cantidad'] = df_ventas_1['cantidad'].fillna(1)

#Convertir a entero ya que las cantidades están como decimales
df_ventas_1['cantidad'] = df_ventas_1['cantidad'].astype(int)

df_ventas_1[df_ventas_1['id_venta'] == df_ventas[df_ventas['cantidad'].isna()]['id_venta'].iloc[0]] 
```

```python
#Productos: Precio nulos -> imputar con el promedio osea la media

categoria_afectada = df_productos_1[df_productos_1['precio'].isna()]['categoria'].iloc[0]

precio_promedio = df_productos_1[df_productos_1['categoria'] == categoria_afectada]['precio'].mean()

df_productos_1['precio'] = df_productos_1['precio'].fillna(precio_promedio)

#Verify
df_productos_1
```

```python
#Clientes -> ciudad nula ------> marca como 'Ciudad Desconocida'

df_clientes_1['ciudad'] = df_clientes_1['ciudad'].fillna('Ciudad Desconocida')

df_clientes_1[df_clientes_1['id_cliente'] == 'C08']
```

## Duplicados

```python
#Buscamos filas duplicadas
df_productos_1[df_productos_1.duplicated(keep=False)] #Keep false = marca todas las ocurrencias para que podamos verlas
```

```python
#Duplicados por id_cliente en clientes
df_clientes_1[df_clientes_1.duplicated(subset=['id_cliente'], keep=False)]
```

```python
#Eliminación de duplicados productos
df_productos_1 = df_productos_1.drop_duplicates(keep='first').reset_index(drop=True)

#Clientes.
df_clientes_1 = df_clientes_1.drop_duplicates(subset=['id_cliente'], keep='first').reset_index(drop=True)

print(f"Productos: {len(df_productos)} -> {len(df_productos_1)} filas")
print(f"Clientes: {len(df_clientes)} -> {len(df_clientes_1)} filas")


```
