## Escalado de Variables:


Un producto = $500

la media es = $229

Hay una difrenecia de $271, el producto está $271 dolares por encima del primedio

std = $450

271 / 450 = 0.60. El producto está aprox 0.6 desviacioones estandar por encima del promedio

z = 0.60

```python
#Seleccinonar las 4 variables necesarias

variables_nums = ['Sales', 'Quantity', 'Discount', 'Profit']

#Creamos dataframe
df_nums = df[variables_nums].copy()

#Visualización de los datos
fig, axes = plt.subplots(1, 4, figsize=(18, 4))

colores = ['steelblue', 'darkorange', 'seagreen', 'tomato']

for i, col in enumerate(variables_nums):
    df_nums[col].plot(kind='hist', bins=40, ax=axes[i], color=colores[i], edgecolor='white')
    
    axes[i].set_title(f'{col}\nRango: [{df_nums[col].min():.1f}, {df_nums[col].max():.1f}]')
    axes[i].set_xlabel('Valor')
    axes[i].set_ylabel('Frecuencia')
    

plt.suptitle('Distribuciones ANTES del Escalado para notar la dierencia de maginitudes', fontsize=13, fontweight='bold', y=1.02)

plt.tight_layout()
plt.show()
```

```python
#Construir una tabla de comparacion entre estadisticos por variable

resumen_magnitudes = pd.DataFrame({
    'Media': df_nums.mean().round(2),
    'Desv: Estándar': df_nums.std().round(2),
    'Minimo': df_nums.min().round(2),
    'Maximo': df_nums.max().round(2),
    'Rango Total': (df_nums.max() - df_nums.min()).round(2)
})

resumen_magnitudes
```

## Estandarización de variables = StandarScaler()

```python
#Paso 1: INstanciar o llamar al Método
#Creación del objeto

scaler = StandardScaler()

#Paso 2: fit_transform
datos_escalados = scaler.fit_transform(df_nums)

#Pase 3: Convertir el arreglo a dataframe
df_escalado = pd.DataFrame(
    datos_escalados,
    columns=[f'{col}_z' for col in variables_nums]
)

df_escalado.head()
```

```python
#Verificación matemática de que el escalado funiconó correctamente

verificacion_scaler = pd.DataFrame({
    'Media (debe ser equivalente a 0)': df_escalado.mean().round(6),
    'Desv. Estandar (debe ser equivalente a 1)': df_escalado.std().round(6)
})

verificacion_scaler
```
